```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_excel('/content/ApexPlanet_DataAnalytics_Dataset.xlsx')

# Display first 5 rows
print(df.head())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Dataset Information
print("\nDataset Info:")
print(df.info())

# Statistical Summary
print("\nStatistical Summary:")
print(df.describe())

# Total Sales
total_sales = df['Total_Sales'].sum()
print("\nTotal Sales:", total_sales)

# Average Age
print("Average Age:", df['Age'].mean())

# Sales by Category
category_sales = df.groupby('Category')['Total_Sales'].sum()
print("\nSales by Category:")
print(category_sales)

# Sales by City
city_sales = df.groupby('City')['Total_Sales'].sum()
print("\nSales by City:")
print(city_sales.sort_values(ascending=False))
```

```python
category_sales.plot(kind='bar', figsize=(8,5))
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
category_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Category-wise Sales")
plt.ylabel("")
plt.show()
```

```python
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()
```
